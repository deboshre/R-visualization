```r
library(rtweet)
c = file.path(paste('tweets_flu_with_statecount', ".csv", sep =""))
mydata = read_twitter_csv(c, unflatten = FALSE)
```

```r
# install.packages("usmap")
library(usmap)
library(ggplot2)

plot_usmap(data = mydata, values = "count",labels=TRUE)+
 scale_fill_continuous(name = "Frequency of flu tweets",low ='yellow',high='firebrick4') + 
  theme(legend.position = "right") + labs(title = "Distribution of tweets related to flu over the states of USA in the year 2019")
```
