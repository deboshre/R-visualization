#### Problem 1: Define two synthetic vectors of data representing sales over 12 months for 2 items. Compare the two using lines graphs.

Discussion: As variable sales2 uses the random generation of 12 numbers with poisson distribution, 
it generates different numbers using the value of lambda(vector of means) as 34. We run it multiple times
and everytime the line graph for sales2 differs.

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(0,80), xlab="Month", ylab="Sales" )
#plot is a generic fucntion for plotting objects
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```
